```python
#Importa as bibliotecas necessarias para a criação dos gráficos
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sn
```

```python
#Salva o dataset na variavel arquivo1 que faz a leitura diretamente do repositorio do github atraves da biblioteca pandas
arquivo1 = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/weight_chart.txt', sep='\t')
age = arquivo1['Age']         #Salva os valores da tabela Age do dataset na variavel age 
weight = arquivo1['Weight']    #Salva os valores da tabela Weight do dataset na variavel weight
```

```python
plt.figure(figsize=(8,6))           #Define o tamanho da imagem do grafico 
plt.plot(age, weight, marker='o')   #Defini o eixo X e o eixo Y alem do marcador dos pontos no plot
plt.yticks([2,4,6,8,10])            #Defini a escala do eixo Y
plt.title("The Relationship between age and weight in a growinfg infant") #Define o titulo do gráfico
plt.xlabel("Age (months)")          #Define o titulo do eixo X do gráfico
plt.ylabel("Weight (kg)")           #Define o titulo do eixo Y do gráfico
plt.show()                          #Exibe o gráfico
```

```python
distribuicao1 = np.random.normal(size=10000)  #Gera um,a distribuição normal atravez da biblioteca Numpy
distribuicao2 = distribuicao1 + 4             #Gera uma segunda distribuição com a junção a 1 distribução gerada com um deslçocamento de + 4
plt.figure(figsize=(10,8))                    #Define o tamanho da imagem do grafico
sn.histplot(np.concatenate((distribuicao1,distribuicao2)), bins=130, color='gray', ec='black') #Defini o eixo X e o eixo Y alem de selecionar a cor do gráfico e sua proporção
plt.title("Mixed distribuition histogram")    #Define o titulo do gráfico
plt.xlabel("Values")                          #Define o titulo do eixo X do gráfico
plt.ylabel("Frequency")                       #Define o titulo do eixo Y do gráfico
plt.xticks([-4, -2, 0, 2, 4, 6, 8])           #Defini a escala do eixo X
plt.yticks([0, 100, 200, 300, 400])           #Defini a escala do eixo Y
plt.show()                                    #Exibe o gráfico
```

```python
#salva o dataset na variavel arquivo7 que faz a leitura diretamente do repositorio do github atraves da biblioteca pandas
arquivo7 = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/brain_bodyweight.txt', sep='\t')
```

```python
plt.figure(figsize=(8,8))   #Define o tamanho da imagem do grafico

#É feito um loop nos valores onde são armazenados em suas respsctivas variaveis cada atributo.
#É feito o calculo do errorbar, juntamente com o caclulo da posição de cada item para sua respectiva apresentação no grafica
for x in arquivo7.values:
  species = x[0]
  bodyweight = x[1]
  brainweight = x[2]
  bodyweightSEM = x[3]
  brainweightSEM = x[4]
  plt.errorbar(y=bodyweight, x=brainweight, yerr=bodyweightSEM, xerr=brainweightSEM, color='black', marker='h' )
  plt.annotate(species, (brainweight - 0.20, bodyweight - 0.3), fontsize = 'small')

plt.xlabel("Brain weight")      #Define o titulo do eixo X do gráfico
plt.ylabel("Body weight")       #Define o titulo do eixo Y do gráfico
plt.xticks([0,1,2,3])           #Defini a escala do eixo X
plt.yticks([-1,0,1,2,3,4,5])    #Defini a escala do eixo Y
plt.show()                      #Exibe o gráfico
```
